```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
a = np.array([1,2,3,4])
plt.figure(figsize=(15,15))
plt.subplot(3,3,1)
for i in range(3):
    for j in range(3):
        plt.subplot(3,3,i*3+j+1)
        if i==j:
            plt.plot(a,a**3)
```

```python
a = plt.figure()
a1 = a.add_axes([0,0,1,1])
a2 = a.add_axes([0.8,0.1,0.1,0.1])
a3 = a.add_axes([0,0.6,0.4,0.4])
a1.plot([1,2,3])
a2.plot([1,2,3])
a3.plot([1,2,3])
```
